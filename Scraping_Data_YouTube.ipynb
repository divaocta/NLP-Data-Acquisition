### **Import Library**

```python
from youtube_comment_downloader import YoutubeCommentDownloader
import pandas as pd
```

### **Input URL Video**

```python
downloader = YoutubeCommentDownloader()

url1 = input("Masukkan URL video 1: ")
url2 = input("Masukkan URL video 2: ")

comments = []
```

### **Scraping Komentar Video Pertama**

```python
for comment in downloader.get_comments_from_url(url1, sort_by=0, language='id'):

    data = {
        "username": comment['author'],
        "komentar": comment['text'],
        "likes": comment['votes'],
        "waktu": comment['time']
    }

    comments.append(data)
```

### **Scraping Komentar Video Kedua**

```python
for comment in downloader.get_comments_from_url(url2, sort_by=0, language='id'):

    data = {
        "username": comment['author'],
        "komentar": comment['text'],
        "likes": comment['votes'],
        "waktu": comment['time']
    }

    comments.append(data)
```

### **Gabungan Data Frame**

```python
df = pd.DataFrame(comments)

print("Total komentar:", len(df))

df.head()
```

### **Simpan File**

```python
df.to_excel("scraping_youtube.xlsx")
```
