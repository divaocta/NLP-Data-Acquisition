### **Import Library**

```python
import pandas as pd
from google_play_scraper import reviews_all, Sort
```

### **Input ID Aplikasi**

```python
app_id = input("Masukkan ID aplikasi: ")

result = reviews_all(
    app_id,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
)
```

### **Scraping Review Aplikasi Jaga by KPK**

```python
data = []

for review in result:

    item = {
        "username": review['userName'],
        "rating": review['score'],
        "ulasan": review['content'],
        "tanggal": review['at']
    }

    data.append(item)

df = pd.DataFrame(data)

print("Jumlah data:", len(df))
df.head()
```

### **Simpan File**

```python
df.to_excel("scraping_googleplay.xlsx")
```
